6) Why Failure in Specific states
-----------------------
Based on the map we can see that there are 3 states that seem to be outliers to the situation those are Georgia, Florida, and Illinios: 

For Georgia there are several reasons why the bank failure was so high: The main reasons are that Atlanta did not had problems with a price bubble, but over construction due to low construction permit requirements and easy to construct land. As, a result when the financial crisis came, banks ended up with a lot of inventory assets (houses) these assets we fixed and hard to convert to cash. In addition, around 5% of Georgiaâ€™s economy was dependent on the construction industry, when the housing market hit Georgia it took a heavier hit than other states. 

Florida: In Florida they suffered a similar fate to Georgia, many small communities in Florida were dependent on the construction industry and since, there were low construction requirements there was a lot of over building. However, the situation was worse on Florida because there was a huge price bubble too, Houses doubled and tripled in their value, which when the market crashed the value of assets went down at a critical rate.  

Illinois: The problem in Illinois was that mortgages were readily available more than other states, when the recession hit, at least 30% of homeowner went under water on their debt, and house value plummet to alarming rates. The main is that recovery in Illinois has been sluggish compared to other states.
